```python
import pandas as pd
```

```python
all_inven = pd.read_csv("rentals/All/rentalInventory_All.csv")
OneBd = pd.read_csv("rentals/OneBd/rentalInventory_OneBd.csv")
Studio = pd.read_csv("rentals/Studio/rentalInventory_Studio.csv")
ThreePlusBd = pd.read_csv("rentals/ThreePlusBd/rentalInventory_ThreePlusBd.csv")
TwoBd = pd.read_csv("rentals/TwoBd/rentalInventory_TwoBd.csv")

all_inven = all_inven[['areaName','Borough','areaType','2018-10']].rename(columns = {'2018-10':'All'})
OneBd = OneBd[['areaName','Borough','areaType','2018-10']].rename(columns = {'2018-10':'OneBd'})
Studio = Studio[['areaName','Borough','areaType','2018-10']].rename(columns = {'2018-10':'Studio'})
ThreePlusBd = ThreePlusBd[['areaName','Borough','areaType','2018-10']].rename(columns = {'2018-10':'ThreePlusBd'})
TwoBd = TwoBd[['areaName','Borough','areaType','2018-10']].rename(columns = {'2018-10':'TwoBd'})

inventory = pd.merge(all_inven,OneBd,on = ['areaName','Borough','areaType'])
inventory = pd.merge(inventory,Studio,on = ['areaName','Borough','areaType'])
inventory = pd.merge(inventory,TwoBd,on = ['areaName','Borough','areaType'])
inventory = pd.merge(inventory,ThreePlusBd,on = ['areaName','Borough','areaType'])
inventory
```

```python
listings = pd.read_csv("listings.csv")
```

```python
for i in listings.columns:
    print(i,'|||',listings[i][0])
```

```python
listings = listings[['neighbourhood_cleansed','neighbourhood_group_cleansed','city', 'state', 'zipcode', 'market','latitude', 'longitude','property_type', 'room_type', 'accommodates',
       'bathrooms', 'bedrooms', 'beds', 'bed_type','price', 'weekly_price', 'monthly_price','has_availability', 'availability_30', 'availability_60',
       'availability_90', 'availability_365']]
```

```python
listings['room_type'].unique()
```

```python
listings = listings[listings['room_type'] == 'Entire home/apt']
```

```python
listings['bedrooms'].value_counts()
```

```python
listings.loc[listings['bedrooms'] > 3,'bedrooms'] = 3
```

```python
listings['bedrooms'].value_counts()
```

```python
listings['price'] = listings['price'].apply(lambda x: float(x.strip('$').replace(',','')))
```

```python
airbnb_all = listings.groupby(['neighbourhood_group_cleansed'])[['price']].agg(['median','count']).reset_index().rename(columns = {'neighbourhood_group_cleansed':'Borough'})
```

```python
listings = listings.groupby(['neighbourhood_group_cleansed','bedrooms'])[['price']].agg(['median','count']).reset_index().rename(columns = {'neighbourhood_group_cleansed':'Borough'})
```

```python
airbnb_all.columns = [' '.join(col).strip() for col in airbnb_all.columns.values]
airbnb_all = airbnb_all.rename(columns = {'price median':'airbnb_price','price count':'airbnb_inventory'})
```

```python
listings.columns = [' '.join(col).strip() for col in listings.columns.values]
listings = listings.rename(columns = {'price median':'airbnb_price','price count':'airbnb_inventory'})
listings
```

```python
listings.loc[:,'bedrooms'] = listings['bedrooms'].map({0:'Studio',1:'OneBd',2:'TwoBd',3:'ThreePlusBd'})
```

```python
airbnb_all['bedrooms'] = 'All'
```

```python
airbnb = pd.concat([airbnb_all,listings])
airbnb
```

```python
price = pd.melt(price, id_vars=['Borough'], value_vars=['All', 'Studio','OneBd','TwoBd','ThreePlusBd']).rename(columns = {'variable':'type','value':'price'})
inventory = pd.melt(inventory, id_vars=['Borough'], value_vars=['All', 'Studio','OneBd','TwoBd','ThreePlusBd']).rename(columns = {'variable':'type','value':'inventory'})
```

```python
rental = price.merge(inventory,on = ['Borough','type'])
```

```python
host = rental.merge(airbnb, left_on = ['Borough','type'], right_on=['Borough','bedrooms'])
```

```python
host = host.drop(['bedrooms'], axis = 1).rename(columns = {'price':'rental_price','inventory':'rental_inventory'})
```

```python
host.to_csv('../data/host.csv')
```
